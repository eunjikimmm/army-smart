```python
## 파일 뒤에 날짜가 있으면 해당 날짜 삭제

import os
import re

def remove_date_from_filename(directory):
    # 정규 표현식 패턴 설정 (예: _20220923)
    date_pattern = re.compile(r'_\d{8}(?=\.[^.]+$)')

    # 디렉터리에 있는 파일들을 순회
    for filename in os.listdir(directory):
        # 정규 표현식으로 날짜 부분이 있는지 확인
        match = date_pattern.search(filename)
        
        if match:
            # 날짜 부분을 제거한 새로운 파일 이름 생성
            new_name = filename[:match.start()] + filename[match.end():]
            old_path = os.path.join(directory, filename)
            new_path = os.path.join(directory, new_name)
            
            # 파일 이름 변경
            os.rename(old_path, new_path)
            print(f"Renamed: {filename} -> {new_name}")
        else:
            print(f"Skipped: {filename}")
            
directory_path = './retrieval_data/candidate_pool_url' ## 폴더 경로 선택
remove_date_from_filename(directory_path)
```

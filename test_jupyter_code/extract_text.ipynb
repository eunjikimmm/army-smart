```python
# json 하나가 여러줄일 경우
def load_jsonl_with_indent(file_path):
    data = []
    # with open(file_path, 'r', encoding='utf-8') as file:
        # json_object = ''
        # inside_object = False

        # for line in file:
        #     if line.strip().startswith('{'):
        #         inside_object = True
        #         json_object = line
        #     elif inside_object:
        #         json_object += line
        #         if line.strip().endswith('}'):
        #             data.append(json.loads(json_object))
        #             inside_object = False
        #             json_object = ''
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data.append(json.loads(line))
    return data

def save_as_jsonl(data, file_path):
    with open(file_path, 'w', encoding='utf-8') as file:
        for item in data:
            json.dump(item, file, ensure_ascii=False)
            file.write('\n')  # 각 JSON 객체 후에 줄바꿈을 추가합니다.

# file_path = './candidate_pool/'+ train_set[0]
```

```python
for set in tqdm(train_set):
    candidate = load_jsonl_with_indent('./candidate_pool/'+ set)
    train_c = [{'text':c['subject']+c['text']} for c in candidate]
    save_as_jsonl(train_c,'./candidate_pool/llm_train_text/'+ set)
```

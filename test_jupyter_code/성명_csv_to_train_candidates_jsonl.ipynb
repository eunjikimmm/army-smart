```python
import json
import pandas as pd
import os
```

```python
# excel 리스트 만들기

work_dir = "./"
excel_dir = "./"
train_jsonl_name = "ncr_digital_train.jsonl"
cand_jsonl_name = "ncr_digital_candi.jsonl"
excel_list = []

for file in os.listdir(excel_dir):
  if file.endswith(".csv"):
    excel_list.append(file)

print(excel_list)
len(excel_list)
```

```python
# convert excel data files to each csv data files

for file in excel_list:
  # excel to csv
  #filename = os.path.splitext(file)[0][:-4] + ".csv"
  filename = os.path.splitext(file)[0] + ".csv"

  xlsx = pd.read_excel(excel_dir + file)
  csv_from_xl = xlsx.to_csv(excel_dir + filename)
```

```python
data_csv = pd.read_csv(excel_dir + file) ## 파일 불러오기
data_csv.rename(columns = {'Question':'source', 'Answer':'target'}, inplace= True)
# data_csv
train_dict_list = data_csv[['source','target']].to_dict(orient='records') ## QA만 뽑아서 json으로 변환
# train_dict_list

# jsonl 파일로 변환(학습 데이터)
with open(work_dir + train_jsonl_name, 'a', encoding='UTF-8') as f:
    for i in train_dict_list:
        #print(i)
        f.write(json.dumps(i, ensure_ascii=False)+ '\n')

# jsonl 파일로 변환(candidate pool)
with open(work_dir + cand_jsonl_name, 'w', encoding='UTF-8') as cand_f:
  with open(work_dir + train_jsonl_name, 'r', encoding='UTF-8') as train_f:
    train_data = train_f.read()
    train_data = train_data.replace('source', 'text')
    cand_data = train_data.replace('target', 'answer')
  cand_f.write(cand_data)
```
